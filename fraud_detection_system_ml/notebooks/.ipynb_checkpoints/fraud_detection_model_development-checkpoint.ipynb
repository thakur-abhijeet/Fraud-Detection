# Fraud Detection System - Model Development

This notebook focuses on:
1. Training multiple models
2. Model evaluation and comparison
3. Hyperparameter tuning
4. Feature importance analysis
5. Model selection and finalization

```python
# Import required libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score,
    confusion_matrix, classification_report, roc_curve, auc,
    precision_recall_curve, average_precision_score
)
from sklearn.model_selection import GridSearchCV, cross_val_score
import joblib

# Set display options
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 100)
```

```python
import joblib

# === Load preprocessed data and preprocessor ===
preprocessed_data = joblib.load('../data/preprocessed_data.joblib')
preprocessor = joblib.load('../models/preprocessor.joblib')

# === Extract raw and processed data ===
X_train = preprocessed_data['X_train']
X_test = preprocessed_data['X_test']
y_train = preprocessed_data['y_train']
y_test = preprocessed_data['y_test']
X_train_resampled = preprocessed_data['X_train_resampled']
y_train_resampled = preprocessed_data['y_train_resampled']

# === Transform data using preprocessor ===
# Only transform raw X_train and X_test. X_train_resampled is already transformed.
X_train_processed = preprocessor.fit_transform(X_train)
X_test_processed = preprocessor.transform(X_test)

# Don't re-transform already processed/resampled data
X_train_resampled_processed = X_train_resampled

# === Optional: Sanity checks ===
print("X_train_processed shape:", X_train_processed.shape)
print("X_test_processed shape:", X_test_processed.shape)
print("X_train_resampled_processed shape:", X_train_resampled_processed.shape)
```

## 1. Baseline Models

```python
def evaluate_model(y_true, y_pred, y_pred_proba=None):
    """Evaluate model performance with various metrics"""
    print("Classification Report:")
    print(classification_report(y_true, y_pred))
    
    print("\nConfusion Matrix:")
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
    plt.title('Confusion Matrix')
    plt.ylabel('True Label')
    plt.xlabel('Predicted Label')
    plt.show()
    
    if y_pred_proba is not None:
        # ROC Curve
        fpr, tpr, _ = roc_curve(y_true, y_pred_proba)
        roc_auc = auc(fpr, tpr)
        
        plt.figure(figsize=(8, 6))
        plt.plot(fpr, tpr, label=f'ROC curve (AUC = {roc_auc:.2f})')
        plt.plot([0, 1], [0, 1], 'k--')
        plt.xlabel('False Positive Rate')
        plt.ylabel('True Positive Rate')
        plt.title('ROC Curve')
        plt.legend()
        plt.show()
        
        # Precision-Recall Curve
        precision, recall, _ = precision_recall_curve(y_true, y_pred_proba)
        avg_precision = average_precision_score(y_true, y_pred_proba)
        
        plt.figure(figsize=(8, 6))
        plt.plot(recall, precision, label=f'PR curve (AP = {avg_precision:.2f})')
        plt.xlabel('Recall')
        plt.ylabel('Precision')
        plt.title('Precision-Recall Curve')
        plt.legend()
        plt.show()
```

```python
# Train and evaluate baseline models
models = {
    'Logistic Regression': LogisticRegression(random_state=42, max_iter=1000),
    'Random Forest': RandomForestClassifier(random_state=42),
    'Gradient Boosting': GradientBoostingClassifier(random_state=42)
}

# Dictionary to store results
results = {}

# Train and evaluate each model
for name, model in models.items():
    print(f"\n{'='*50}")
    print(f"Training and evaluating {name}")
    print(f"{'='*50}")
    
    # Train model on resampled data
    model.fit(X_train_resampled_processed, y_train_resampled)
    
    # Make predictions
    y_pred = model.predict(X_test_processed)
    y_pred_proba = model.predict_proba(X_test_processed)[:, 1]
    
    # Evaluate model
    evaluate_model(y_test, y_pred, y_pred_proba)
    
    # Store results
    results[name] = {
        'model': model,
        'accuracy': accuracy_score(y_test, y_pred),
        'precision': precision_score(y_test, y_pred),
        'recall': recall_score(y_test, y_pred),
        'f1': f1_score(y_test, y_pred)
    }

# Compare model performances
print("\nModel Performance Comparison:")
comparison_df = pd.DataFrame(results).T
comparison_df = comparison_df.drop('model', axis=1)
print(comparison_df.round(4))

# Visualize model comparison
plt.figure(figsize=(10, 6))
comparison_df.plot(kind='bar', ax=plt.gca())
plt.title('Model Performance Comparison')
plt.xlabel('Models')
plt.ylabel('Score')
plt.xticks(rotation=45)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
# Save all models and results
import os
from datetime import datetime

# Create models directory if it doesn't exist
models_dir = '../models/baseline'
os.makedirs(models_dir, exist_ok=True)

# Save timestamp for versioning
timestamp = datetime.now().strftime('%Y%m%d_%H%M%S')

# Save each model
for name, result in results.items():
    model = result['model']
    model_path = os.path.join(models_dir, f'{name.lower().replace(" ", "_")}_{timestamp}.joblib')
    joblib.dump(model, model_path)
    print(f"Saved {name} to {model_path}")

# Save comparison results
results_path = os.path.join(models_dir, f'model_comparison_{timestamp}.csv')
comparison_df.to_csv(results_path)
print(f"\nSaved model comparison results to {results_path}")

# Save the best model based on F1 score
best_model_name = comparison_df['f1'].idxmax()
best_model = results[best_model_name]['model']
best_model_path = os.path.join(models_dir, f'best_model_{timestamp}.joblib')
joblib.dump(best_model, best_model_path)
print(f"\nBest model ({best_model_name}) saved to {best_model_path}")

# Print best model metrics
print("\nBest Model Metrics:")
print(comparison_df.loc[best_model_name])
```

```python
# =============================================
# Hyperparameter Tuning
# =============================================

print("\n" + "="*50)
print("Hyperparameter Tuning")
print("="*50)

# Define parameter grids for each model type
param_grids = {
    'Logistic Regression': {
        'C': [0.001, 0.01, 0.1, 1, 10, 100],
        'penalty': ['l1', 'l2'],
        'solver': ['liblinear', 'saga']
    },
    'Random Forest': {
        'n_estimators': [100, 200, 300],
        'max_depth': [10, 20, 30, None],
        'min_samples_split': [2, 5, 10],
        'min_samples_leaf': [1, 2, 4]
    },
    'Gradient Boosting': {
        'n_estimators': [100, 200, 300],
        'learning_rate': [0.01, 0.1, 0.2],
        'max_depth': [3, 4, 5],
        'min_samples_split': [2, 5, 10]
    }
}

# Get the best model name from previous results
best_model_name = comparison_df['f1'].idxmax()
print(f"\nPerforming hyperparameter tuning for {best_model_name}")

# Initialize the model
if best_model_name == 'Logistic Regression':
    base_model = LogisticRegression(random_state=42, max_iter=1000)
elif best_model_name == 'Random Forest':
    base_model = RandomForestClassifier(random_state=42)
else:  # Gradient Boosting
    base_model = GradientBoostingClassifier(random_state=42)

# Perform GridSearchCV
grid_search = GridSearchCV(
    estimator=base_model,
    param_grid=param_grids[best_model_name],
    cv=5,
    scoring='f1',
    n_jobs=-1,
    verbose=2
)

# Fit the grid search
print("\nStarting GridSearchCV...")
grid_search.fit(X_train_resampled_processed, y_train_resampled)

# Print best parameters and score
print("\nBest parameters:", grid_search.best_params_)
print("Best cross-validation score:", grid_search.best_score_)

# Get the best model
best_tuned_model = grid_search.best_estimator_

# Evaluate the tuned model
print("\nEvaluating tuned model on test set...")
y_pred_tuned = best_tuned_model.predict(X_test_processed)
y_pred_proba_tuned = best_tuned_model.predict_proba(X_test_processed)[:, 1]

# Compare with original model
print("\nOriginal model metrics:")
print(comparison_df.loc[best_model_name])
print("\nTuned model metrics:")
print(f"Accuracy: {accuracy_score(y_test, y_pred_tuned):.4f}")
print(f"Precision: {precision_score(y_test, y_pred_tuned):.4f}")
print(f"Recall: {recall_score(y_test, y_pred_tuned):.4f}")
print(f"F1 Score: {f1_score(y_test, y_pred_tuned):.4f}")

# Save the tuned model
tuned_model_path = os.path.join(models_dir, f'tuned_{best_model_name.lower().replace(" ", "_")}_{timestamp}.joblib')
joblib.dump(best_tuned_model, tuned_model_path)
print(f"\nSaved tuned model to {tuned_model_path}")

# Save the grid search results
grid_search_results = pd.DataFrame(grid_search.cv_results_)
grid_search_results_path = os.path.join(models_dir, f'grid_search_results_{timestamp}.csv')
grid_search_results.to_csv(grid_search_results_path)
print(f"Saved grid search results to {grid_search_results_path}")

# Visualize the impact of different parameters
plt.figure(figsize=(12, 6))
for param in grid_search.best_params_.keys():
    param_values = grid_search_results[f'param_{param}'].astype(str)
    mean_scores = grid_search_results['mean_test_score']
    
    plt.subplot(1, len(grid_search.best_params_), list(grid_search.best_params_.keys()).index(param) + 1)
    plt.scatter(param_values, mean_scores)
    plt.title(f'Impact of {param}')
    plt.xticks(rotation=45)
    plt.ylabel('Mean CV Score')

plt.tight_layout()
plt.show()
```

```python
# =============================================
# Feature Importance Analysis
# =============================================

print("\n" + "="*50)
print("Feature Importance Analysis")
print("="*50)

# Fit the preprocessor if not already fitted
if not hasattr(preprocessor, 'transformers_'):
    print("Fitting preprocessor...")
    preprocessor.fit(X_train)

# Get feature names from the preprocessor
feature_names = preprocessor.get_feature_names_out()
print(f"\nNumber of features: {len(feature_names)}")

# Function to plot feature importance
def plot_feature_importance(importance, feature_names, top_n=20):
    # Create a DataFrame for easier plotting
    importance_df = pd.DataFrame({
        'Feature': feature_names,
        'Importance': importance
    })
    
    # Sort by importance
    importance_df = importance_df.sort_values('Importance', ascending=False)
    
    # Plot top N features
    plt.figure(figsize=(12, 6))
    sns.barplot(x='Importance', y='Feature', data=importance_df.head(top_n))
    plt.title(f'Top {top_n} Most Important Features')
    plt.tight_layout()
    plt.show()
    
    return importance_df

# Get feature importance based on model type
if isinstance(best_tuned_model, LogisticRegression):
    # For Logistic Regression, use absolute coefficients
    importance = np.abs(best_tuned_model.coef_[0])
elif isinstance(best_tuned_model, (RandomForestClassifier, GradientBoostingClassifier)):
    # For tree-based models, use feature_importances_
    importance = best_tuned_model.feature_importances_
else:
    raise ValueError("Model type not supported for feature importance analysis")

# Plot and get feature importance
importance_df = plot_feature_importance(importance, feature_names)

# Print top 10 most important features
print("\nTop 10 Most Important Features:")
print(importance_df.head(10))

# Analyze feature importance distribution
plt.figure(figsize=(10, 6))
plt.hist(importance, bins=50)
plt.title('Distribution of Feature Importance')
plt.xlabel('Importance')
plt.ylabel('Count')
plt.show()

# Calculate cumulative importance
importance_df['Cumulative Importance'] = importance_df['Importance'].cumsum()
importance_df['Cumulative Percentage'] = importance_df['Cumulative Importance'] / importance_df['Importance'].sum() * 100

# Plot cumulative importance
plt.figure(figsize=(10, 6))
plt.plot(range(len(importance_df)), importance_df['Cumulative Percentage'])
plt.title('Cumulative Feature Importance')
plt.xlabel('Number of Features')
plt.ylabel('Cumulative Importance (%)')
plt.grid(True)
plt.show()

# Find number of features needed for 95% importance
features_95 = importance_df[importance_df['Cumulative Percentage'] <= 95].shape[0]
print(f"\nNumber of features needed for 95% importance: {features_95}")

# Save feature importance results
feature_importance_path = os.path.join(models_dir, f'feature_importance_{timestamp}.csv')
importance_df.to_csv(feature_importance_path)
print(f"\nSaved feature importance results to {feature_importance_path}")
```
